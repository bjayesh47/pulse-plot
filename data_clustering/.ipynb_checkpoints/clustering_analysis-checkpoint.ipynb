```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
from matplotlib import rcParams
import numpy as np
```

```python
df = pd.read_csv('../feature_complete_final_data.csv')
df.columns
```

```python
df = df.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0', 'Unnamed: 0_x',
                        'Unnamed: 0_y', 'type', 'uri', 'track_href',
                        'analysis_url', 'duration_ms', 'time_signature',
                        'end_time', 'ms_played'])
```

```python
df_analysis = df[['danceability', 'energy', 'loudness',	'speechiness', 'acousticness',	'instrumentalness',	'liveness',	'valence',	'tempo'	]]
df_analysis.head()
```

```python
from sklearn import preprocessing
```

```python
df_analysis_scaled = preprocessing.scale(df_analysis)
```

```python
from sklearn.manifold import TSNE
```

```python
df_tsne = TSNE(learning_rate = 100).fit_transform(df_analysis_scaled)
df_tsne
```

```python
rcParams['figure.figsize'] = 20, 8
sns.scatterplot(x = df_tsne[:, 0], y = df_tsne[:, 1])
```

```python
data = pd.DataFrame({'x': df_tsne[:, 0], 'y': df_tsne[:, 1]})
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1, 15):
    kmeans = KMeans(i)
    kmeans.fit(df_analysis_scaled)
    wcss.append(kmeans.inertia_)
wcss
```

```python
sns.lineplot(data = (range(1, 15), wcss))
plt.xlabel('Label Count')
plt.ylabel('WCSS')
```

```python
kmeans_model = KMeans(6)
kmeans_model.fit(df_analysis_scaled)
data['kmeans'] = kmeans_model.fit_predict(df_analysis_scaled)
```

```python
rcParams['figure.figsize'] = 18,8
sns.scatterplot(data = data, x = 'x', y = 'y', hue = 'kmeans', palette = 'viridis')
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model_agg = AgglomerativeClustering(n_clusters = 6)
model_agg.fit(df_analysis_scaled)
data['agg'] = model_agg.fit_predict(df_analysis_scaled)
```

```python
rcParams['figure.figsize'] = 18,8
sns.scatterplot(data = data, x = 'x', y = 'y', hue = 'agg', palette = 'viridis')
```

```python
from sklearn.cluster import DBSCAN
```

```python
model_dbscan = DBSCAN(eps = 0.3, min_samples = 9)
data['dbscan'] = model_dbscan.fit_predict(df_analysis_scaled)
```

```python
rcParams['figure.figsize'] = 18,8
sns.scatterplot(data=data, x='x', y='y', hue='dbscan', palette='viridis')
```
