```python
import json
import pandas as pd
from pprint import pprint
```

```python
with open('../data/StreamingHistory_music_0.json', encoding='utf-8') as f:
    data = json.load(f)
pprint(data[:5])
```

```python
spotify_data = pd.DataFrame()
def extract_json_value(column_name):
    return [i[column_name] for i in data]

spotify_data['artist_name'] = extract_json_value('artistName')
spotify_data['end_time'] = extract_json_value('endTime')
spotify_data['ms_played'] = extract_json_value('msPlayed')
spotify_data['name'] = extract_json_value('trackName')
```

```python
spotify_data.head()
```

```python
final_data = pd.read_csv('./../final_data.csv')
```

```python
analysis_df = pd.merge(final_data, spotify_data, on='name')
analysis_df.head()
```

```python
analysis_df.columns
```

```python
analysis_df = analysis_df.drop(columns = ['Unnamed: 0', 'Unnamed: 0_x', 'Unnamed: 0_y', 'type', 'time_signature', 'uri', 'track_href', 'analysis_url'])
analysis_df.head()
```

```python
analysis_df.to_csv('../song_streaming_history.csv')
```
