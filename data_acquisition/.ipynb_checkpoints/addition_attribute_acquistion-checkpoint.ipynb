```python
import json
import pandas as pd
from pprint import pprint
```

```python
with open('../data/StreamingHistory_music_0.json', encoding='utf-8') as f:
    data = json.load(f)
pprint(data[:5])
```

```python
spotify_data = pd.DataFrame()
def extract_json_value(column_name):
    return [i[column_name] for i in data]

spotify_data['artist_name'] = extract_json_value('artistName')
spotify_data['end_time'] = extract_json_value('endTime')
spotify_data['ms_played'] = extract_json_value('msPlayed')
spotify_data['name'] = extract_json_value('trackName')
```

```python
spotify_data.head(10)
```

```python
spotify_data.shape
```

```python
agg_functions = {'end_time': 'max', 'ms_played': 'sum', 'artist_name': 'first'}
unique_spotify_data = spotify_data.groupby(spotify_data['name']).aggregate(agg_functions)
pprint(unique_spotify_data)
```

```python
final_data = pd.read_csv('./../final_data.csv')
```

```python
result_df = pd.merge(final_data, unique_spotify_data, on='name')
result_df.head()
```

```python
result_df.to_csv('../feature_complete_final_data.csv')
```
